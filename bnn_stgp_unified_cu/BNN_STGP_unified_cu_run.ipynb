```python
sample_size = 2000
data_type = "simulation"
seed = 1
true_r2 = 0.80
gamma = 0.1
in_feature_list = [50, 100]
a_list = [0.0001, 0.0001]
b_list = [50, 50]
hidden_unit_list = [1, 128]
step_gamma = 0.5
step_decay_epoch = 100
lr = 5e-6
epochs = 200
poly_degree = 30
```

```python
all_experiments = []
print(f"seed {seed}")
experiment = Experiment()
trainer = experiment.run_whole_experiment(
    sample_size=sample_size,
    data_type=data_type,
    seed=seed,
    true_r2=true_r2,
    gamma=gamma,
    in_feature_list=in_feature_list,
    a_list=a_list,
    b_list=b_list,
    hidden_unit_list=hidden_unit_list,
    step_gamma=step_gamma,
    step_decay_epoch=step_decay_epoch,
    lr=lr,
    epochs=epochs,
    poly_degree=poly_degree,
)
```

```python
result = experiment.process_results(trainer, burnin_ratio=0.5, gamma=gamma)
all_experiments.append(experiment)
```
